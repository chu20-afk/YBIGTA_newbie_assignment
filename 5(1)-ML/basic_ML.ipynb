# 1. 데이터 로드 및 기본 탐색

```python
import pandas as pd
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Class"].value_counts()
```

```python
df["Class"].value_counts(normalize=True)
```

# 2. 샘플링

```python
# (1) 정상(0), 사기(1) 데이터를 분리
df_fraud = df[df["Class"] == 1]   
df_normal = df[df["Class"] == 0]  

print(len(df_fraud))
print(len(df_normal))
```

```python
# (2) 사기는 전부 유지, 정상 거래는 10,000건만 무작위로 추출 (random_state=42)
df_normal_sampled = df_normal.sample(n=10000, random_state=42)

print(len(df_fraud))
print(len(df_normal_sampled))
```

```python
# (3) 두 데이터를 합쳐서 데이터프레임 만듦.
# 정상(샘플 10,000건) + 사기(전체) 합치기
df_sampled = pd.concat([df_normal_sampled, df_fraud], axis=0)

# 섞기(모델 학습 시 순서 편향 방지)
df_sampled = df_sampled.sample(frac=1, random_state=42).reset_index(drop=True)

print(df_sampled.shape)
```

```python
# (4) 샘플링 후 Class 비율 다시 출력함.

# 샘플링 전 비율
print(df_sampled["Class"].value_counts())

print("\n샘플링 후 Class 비율:")
print(df_sampled["Class"].value_counts(normalize=True))
```

# 3. 데이터 전처리

```python
from sklearn.preprocessing import StandardScaler

# 1) Amount만 표준화해서 Amount_Scaled로 대체
scaler = StandardScaler()
df_sampled["Amount_Scaled"] = scaler.fit_transform(df_sampled[["Amount"]])

# 2) 원본 Amount 제거
df_sampled = df_sampled.drop(columns=["Amount"])

# 3) X(특징), y(정답) 분리
X = df_sampled.drop(columns=["Class"])
y = df_sampled["Class"]

print(X.shape)
print(y.shape)
```

# 4. 학습 데이터와 테스트 데이터 분할

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,          # 8:2 분할 (test 20%)
    stratify=y,             # 클래스 비율 유지
    random_state=42
)
```

```python

# 학습셋 크기, 테스트셋 크기
print(X_train.shape, y_train.shape)
print( X_test.shape, y_test.shape)
```

```python
# 학습셋 클래스 분포 확인
print(y_train.value_counts())
print(y_train.value_counts(normalize=True))
```

```python
# 테스트셋 클래스 분포 확인
print(y_test.value_counts())
print(y_test.value_counts(normalize=True))
```

# 5. SMOTE 적용

```python
# (1) SMOTE 적용 전, 사기 거래 건수 확인
print(y_train.value_counts())
```

```python
# (2) SMOTE 적용 (X_train에만)

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# (3) SMOTE 적용 후, 사기 거래 건수 확인

print(y_train_smote.value_counts())

print(X_train.shape)
print( X_train_smote.shape)
```

# 6. 모델 학습

```python
# (1) 모델 선택: Logistic Regression (baseline)
from sklearn.linear_model import LogisticRegression

# 로지스틱 회귀 모델 (확률 예측)
model = LogisticRegression(max_iter=2000, random_state=42) #학습 2000번 반복

# 학습 (SMOTE 적용된 train으로)
model.fit(X_train_smote, y_train_smote)
```

```python
# (2) 테스트셋 예측값(predict), 예측확률(predict_proba) 출력

# 예측 라벨(0/1)
y_pred = model.predict(X_test)

# 예측 확률(사기일 확률 = class 1 확률)
y_proba = model.predict_proba(X_test)[:, 1]

# predict, predict_proba 결과 앞 20개 확인
print(y_pred[:20])
print( y_proba[:20])
```

```python
# (3) classification_report로 Precision / Recall / F1-score 확인

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred, digits=4))
```

```python
# (4) PR-AUC 계산: average_precision_score

from sklearn.metrics import average_precision_score

pr_auc = average_precision_score(y_test, y_proba)
print(pr_auc)
```

# 7. 최종 성능 평가

```python
# (1) threshold = 0.4
t = 0.4
y_pred = (y_proba >= t).astype(int)
print(classification_report(y_test, y_pred, digits=4))
```

```python
# (2) threshold = 0.3
t = 0.3
y_pred = (y_proba >= t).astype(int)
print(classification_report(y_test, y_pred, digits=4))
```

```python
# (3) C=0.1
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

model = LogisticRegression(C=0.1, max_iter=2000, random_state=42)
model.fit(X_train_smote, y_train_smote)

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]   # class 1(사기) 확률

print(classification_report(y_test, y_pred, digits=4))

pr_auc = average_precision_score(y_test, y_proba)
print("PR-AUC:", round(pr_auc, 4))
```

```python
# C=0.1 맞는지 확인
print("C used =", model.get_params()["C"])
```

```python
# (4) C=1

model = LogisticRegression(C=1, max_iter=2000, random_state=42)
model.fit(X_train_smote, y_train_smote)

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]   # class 1(사기) 확률

print(classification_report(y_test, y_pred, digits=4))

pr_auc = average_precision_score(y_test, y_proba)
print("PR-AUC:", round(pr_auc, 4))
```

```python
# C=1 맞는지 확인
print("C used =", model.get_params()["C"])
```

```python
# (5) C=10

model = LogisticRegression(C=10, max_iter=2000, random_state=42)
model.fit(X_train_smote, y_train_smote)

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]   # class 1(사기) 확률

print(classification_report(y_test, y_pred, digits=4))

pr_auc = average_precision_score(y_test, y_proba)
print("PR-AUC:", round(pr_auc, 4))
```

```python
# # C=10 맞는지 확인
print("C used =", model.get_params()["C"])
```
